### 1. Die Datenmenge, also das Modulhandbuch in ein bearbeitbares Format konventieren. Hier die Word-Datei in eine .txt Datei umwandeln. Dies kann man manuel machen, in dem man beim Speichen eintsprechende Einstellungsparameter mitgibt.

```python
def split_text_file(filename):
    encodings = ['utf-8', 'latin-1', 'iso-8859-1', 'cp1252']  # Liste gängiger Codecs

    for encoding in encodings:
        try:
            with open('modulhandbuch.txt', 'r', encoding=encoding) as file:
                content = file.read()
            break  # Wenn das Lesen erfolgreich war, brechen wir die Schleife ab
        except UnicodeDecodeError:
            continue  # Wenn ein Fehler auftritt, versuchen wir den nächsten Codec

    # Split the content by multiple newlines
    paragraphs = content.split('\n\n\n\n\nB')
    
    # Remove leading and trailing whitespace from each paragraph
    paragraphs = [para.strip() for para in paragraphs]

    return paragraphs

# Usage example
filename = 'your_text_file.txt'
paragraphs = split_text_file(filename)

# Print each paragraph
for i, para in enumerate(paragraphs, start=1):
    print(f"Paragraph {i}:\n{para}\n")
```
